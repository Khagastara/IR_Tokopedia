```python
from tokopaedi import search
import pandas as pd
import os
import time

csv_file = "produk_tokopedia.csv"

keywords = [
    "laptop", "smartphone", "kopi", "sepatu", "baju",
    "kulkas", "kipas angin", "buku", "tas", "headphone",
    "mouse", "keyboard", "monitor", "kamera", "parfum"
]

for keyword in keywords:
    print(f"Scraping: {keyword}...")

    try:
        results = search(keyword, max_result=100)

        data = []
        for item in results:
            d = item.json()
            data.append({
                "keyword":             keyword,
                "product_name":        d.get("product_name", ""),
                "price":               d.get("price", ""),
                "price_original":      d.get("price_original", ""),
                "discount_percentage": d.get("discount_percentage", ""),
                "rating":              d.get("rating", ""),
                "shop.name":           d.get("shop", {}).get("name", ""),
                "url":                 d.get("url", ""),
            })
            
            try:
                price = d.get("price", 0) or 0
                price_original_str = d.get("price_original", "") or ""
                price_original = int(price_original_str.replace("Rp", "").replace(".", "").strip())
                discount = round((1 - price / price_original) * 100) if price_original > 0 else 0
            except:
                discount = 0

            data[-1]["discount_percentage"] = discount

        df = pd.DataFrame(data)
        header = not os.path.exists(csv_file)
        df.to_csv(csv_file, mode="a", index=False, header=header)
        print(f"{len(df)} produk dari '{keyword}' disimpan")

    except Exception as e:
        print(f"'{keyword}' gagal: {e}, lanjut keyword berikutnya...")

    time.sleep(3)

print("\nSelesai. Total data:")
print(pd.read_csv(csv_file)["keyword"].value_counts())
```
